```python
import matplotlib.pyplot as sade
import pandas as pd
```

```python
harj2 = pd.read_csv('Sade_Katinen.csv')
harj2.head(15)
```

```python
paiva = harj2['Pv']
kk = harj2['Kk']
vuosi = harj2['Vuosi']
df = pd.DataFrame({'day': paiva, 'month':kk, 'year':vuosi})
df["date"] = pd.to_datetime(df)
print(df)
ylin = harj2['Ylin lämpötila (degC)']
alin = harj2['Alin lämpötila (degC)']
```

```python
sade.figure(figsize=(20,5)) #geeksforgeeks.org haku pyplot parametres
sade.plot(df["date"], ylin)   #tehdään kuvaaja ylin lämpötila
sade.plot(df["date"], alin)   #kuvaaja alin lämpötila
sade.xlabel('Päivämäärä')
sade.ylabel('Lämpötila')
sade.grid(True)
sade.show()  #Jättää tekniset kommentit pois
```
